```python
import cv2
from PIL import Image
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bicicleta = cv2.imread('Imagenes/Bicicleta.jpg')
bus = cv2.imread('Imagenes/Bus.jpg')
carro = cv2.imread('Imagenes/Carro.jpg')
pare = cv2.imread("Imagenes/Pare.jpg")
planta = cv2.imread('Imagenes/Planta.jpg')
```

```python
plt.imshow(bus)
```

```python
N_kernel = np.array([[-1, -1, -1],[1, -2, 1],[1, 1, 1]])
S_kernel = np.array([[1,1,1],[1,-2,1],[-1,-1,-1]])
E_kernel = np.array([[1,1,-1],[1,-2,-1],[1,1,-1]])
O_kernel = np.array([[-1,1,1],[-1,-2,1],[-1,1,1]])
So_kernel = np.array([[1,1,1],[-1,-2,1],[-1,-1,1]])
No_kernel = np.array([[-1,-1,1],[-1,-2,1],[1,1,1]])
Se_kernel = np.array([[1,1,1],[1,-2,-1],[1,-1,-1]])
Ne_kernel = np.array([[1,-1,-1],[1,-2,-1],[1,1,1]])
To_kernel = N_kernel + S_kernel + E_kernel + O_kernel + So_kernel + No_kernel + Se_kernel + Ne_kernel
```

```python
def Filtro_Prewitt(image):
    plt.figure(1)
    plt.subplots_adjust(top=1.75,bottom=0.3,left=0.10,right=2,hspace=0.25,wspace=0.5)
    imagen_Ne = cv2.filter2D(src=image,ddepth=-1,kernel=Ne_kernel)
    plt.subplot(333),plt.imshow(imagen_Ne)
    plt.title('Filtro direccion Ne'), plt.xticks([]), plt.yticks([])
    imagen_N = cv2.filter2D(src=image,ddepth=-1,kernel=N_kernel)
    plt.subplot(332),plt.imshow(imagen_N)
    plt.title('Filtro direccion N'), plt.xticks([]), plt.yticks([])
    imagen_se = cv2.filter2D(src=image,ddepth=-1,kernel=Se_kernel)
    plt.subplot(339),plt.imshow(imagen_se)
    plt.title('Filtro direccion Se'), plt.xticks([]), plt.yticks([])
    imagen_So = cv2.filter2D(src=image,ddepth=-1,kernel=So_kernel)
    plt.subplot(337),plt.imshow(imagen_So)
    plt.title('Filtro direccion So'), plt.xticks([]), plt.yticks([])
    imagen_S = cv2.filter2D(src=image,ddepth=-1,kernel=S_kernel)
    plt.subplot(338),plt.imshow(imagen_S)
    plt.title('Filtro direccion S'), plt.xticks([]), plt.yticks([])    
    imagen_No = cv2.filter2D(src=image,ddepth=-1,kernel=No_kernel)
    plt.subplot(331),plt.imshow(imagen_No)
    plt.title('Filtro direccion No'), plt.xticks([]), plt.yticks([])
    imagen_O = cv2.filter2D(src=image,ddepth=-1,kernel=O_kernel)
    plt.subplot(334),plt.imshow(imagen_O)
    plt.title('Filtro direccion O'), plt.xticks([]), plt.yticks([])
    imagen_E = cv2.filter2D(src=image,ddepth=-1,kernel=E_kernel)
    plt.subplot(336),plt.imshow(imagen_E)
    plt.title('Filtro direccion E'), plt.xticks([]), plt.yticks([])
    imagen_filtro = cv2.filter2D(src=image,ddepth=-1,kernel=To_kernel)
    ##plt.imshow(imagen_filtro)
    plt.subplot(335),plt.imshow(imagen_filtro)
    plt.title('Filtro total'), plt.xticks([]), plt.yticks([])
    
```

```python
Filtro_Prewitt(bus)
```

```python
Filtro_Prewitt(pare)
```

## Filtro Canny

```python
def filtro_canny (image,inf,sup):
    imagen_filtro = cv2.Canny(image,inf,sup)
    plt.subplot(121),plt.imshow(image,cmap = 'gray')
    plt.title('Imagen original'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(imagen_filtro,cmap = 'gray')
    plt.title('Filtro Canny'), plt.xticks([]), plt.yticks([])
```

```python
filtro_canny(pare,75,255)
```

## Filtro Laplaciano

```python
def filtro_laplaciano (image,depth,kernel_size):
    imagen_filtro = cv2.Laplacian(image,depth,kernel_size)
    plt.subplot(121),plt.imshow(image,cmap = 'gray')
    plt.title('Imagen original'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(imagen_filtro,cmap = 'gray')
    plt.title('Filtro Canny'), plt.xticks([]), plt.yticks([])
```

```python
filtro_laplaciano(carro,-1,10)
```
