En este taller se comparan los histogramas de tres imagenes a blanco y negro de un objeto tomadas en diferentes condiciones de luminosidad. Estas comparaciónes de similitud se realizaron con cuatro métodos: correlación, chi-cuadrado, intersección y distancia de Bhattacharyya. Además se realizó esta comparación para histogramas de la misma imagen generados con un número diferente de bins.


```python
import cv2
from PIL import Image
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sol = cv2.imread("TallerSol.jpg",0)
sol = cv2.resize(sol,(512,512))
sombra = cv2.imread("Taller1Sombra.jpg",0)
sombra = cv2.resize(sombra,(512,512))
mixto = cv2.imread("Taller1Mixto.jpg",0)
mixto = cv2.resize(mixto,(512,512))
```

```python
plt.imshow(sol,cmap='Greys')
plt.title("Sol")
```

```python
plt.imshow(sombra,cmap='Greys')
plt.title("Sombra")
```

```python
plt.imshow(mixto,cmap='Greys')
plt.title("Mixto")
```

## Histograma con 256 bins

```python
hist1 = cv2.calcHist([sol],[0],None,[256],[0,256])
hist2 = cv2.calcHist([sombra],[0],None,[256],[0,256])
hist3 = cv2.calcHist([mixto],[0],None,[256],[0,256])
```

```python
hist1 = cv2.normalize(hist1, hist1, 0, 1, cv2.NORM_MINMAX)
hist2 = cv2.normalize(hist2, hist2, 0, 1, cv2.NORM_MINMAX)
hist3 = cv2.normalize(hist3, hist3, 0, 1, cv2.NORM_MINMAX)
```

```python
plt.plot(hist1), plt.title('Histograma')
plt.plot(hist2), plt.title('Histograma')
plt.plot(hist3), plt.title('Histograma')
plt.legend(['Sol','Sombra','Mixto'])
```

### Resultados

```python
results = np.zeros((4,3))

results[0,0] = cv2.compareHist(hist1,hist2,cv2.HISTCMP_CORREL)
results[1,0] = cv2.compareHist(hist1,hist2,cv2.HISTCMP_CHISQR)
results[2,0] = cv2.compareHist(hist1,hist2,cv2.HISTCMP_INTERSECT)
results[3,0] = cv2.compareHist(hist1,hist2,cv2.HISTCMP_BHATTACHARYYA)

results[0,1] = cv2.compareHist(hist1,hist3,cv2.HISTCMP_CORREL)
results[1,1] = cv2.compareHist(hist1,hist3,cv2.HISTCMP_CHISQR)
results[2,1] = cv2.compareHist(hist1,hist3,cv2.HISTCMP_INTERSECT)
results[3,1] = cv2.compareHist(hist1,hist3,cv2.HISTCMP_BHATTACHARYYA)

results[0,2] = cv2.compareHist(hist2,hist3,cv2.HISTCMP_CORREL)
results[1,2] = cv2.compareHist(hist2,hist3,cv2.HISTCMP_CHISQR)
results[2,2] = cv2.compareHist(hist2,hist3,cv2.HISTCMP_INTERSECT)
results[3,2] = cv2.compareHist(hist2,hist3,cv2.HISTCMP_BHATTACHARYYA)

pd.DataFrame(results, columns=["Sol-Sombra", "Sol-Mixto", "Sombra-Mixto"], index=["Correlation","Chi-Squared","Intersection","Bhattacharyya"])
```

### Descripción
Según el índice de correlación entre las imágenes Sol-Sombra y Sol-Mixto no existe correlación y estas presentan divergencia, aunque no máxima. Con respecto a las imágenes Sombra-Mixto se observa que existe un grado de correlación considerable entre las mismas.
El índice de Chi-cuadrado nos informa un comportamiento similar, las imágenes Sombra-Mixto cuentan con un valor significativamente menor a las otras dos comparaciones. Sin embargo este coeficiente nos indica que existe una divergencia mayor entre las imagenes Sol-Mixto.
Al comparar por intersección se observa el mismo comportamiento que con el indicador chi-cuadrado, en este caso valores altos indican mayor similitud.
La distancia de Bhattacharyya muestra un valor menor en la comparación Sombra-Mixto. Sin embargo, no se pudo calcular el coeficiente para las imagenes Sol-Sombra, esto puede ser debido a la normalización que podría generar valores de cero en el denominador de la ecuación.
