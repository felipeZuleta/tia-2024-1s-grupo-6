```python
import cv2
from PIL import Image
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sol = cv2.imread("TallerSol.jpg",0)
sol = cv2.resize(sol,(512,512))
sombra = cv2.imread("Taller1Sombra.jpg",0)
sombra = cv2.resize(sombra,(512,512))
mixto = cv2.imread("Taller1Mixto.jpg",0)
mixto = cv2.resize(mixto,(512,512))
```

```python
cv2.imshow('Gris',sol) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Histograma con 256 bins

```python
hist1 = cv2.calcHist([sol],[0],None,[256],[0,256])
hist2 = cv2.calcHist([sombra],[0],None,[256],[0,256])
hist3 = cv2.calcHist([mixto],[0],None,[256],[0,256])
```

```python
plt.plot(hist1), plt.title('Histograma')
plt.plot(hist2), plt.title('Histograma')
plt.plot(hist3), plt.title('Histograma')
plt.legend(['Sol','Sombra','Mixto'])
```

```python
cv2.compareHist(hist1,hist2,cv2.HISTCMP_CORREL)
```

```python
cv2.compareHist(hist1,hist2,cv2.HISTCMP_CHISQR)
```

```python
cv2.compareHist(hist1,hist2,cv2.HISTCMP_INTERSECT)
```

```python
cv2.compareHist(hist1,hist2,cv2.HISTCMP_BHATTACHARYYA)
```
