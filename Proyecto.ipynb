```python
import os
import cv2
from PIL import Image
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

import random
```

### Filtros

```python
N_kernel = np.array([[-1, -1, -1],[1, -2, 1],[1, 1, 1]])
S_kernel = np.array([[1,1,1],[1,-2,1],[-1,-1,-1]])
E_kernel = np.array([[1,1,-1],[1,-2,-1],[1,1,-1]])
O_kernel = np.array([[-1,1,1],[-1,-2,1],[-1,1,1]])
So_kernel = np.array([[1,1,1],[-1,-2,1],[-1,-1,1]])
No_kernel = np.array([[-1,-1,1],[-1,-2,1],[1,1,1]])
Se_kernel = np.array([[1,1,1],[1,-2,-1],[1,-1,-1]])
Ne_kernel = np.array([[1,-1,-1],[1,-2,-1],[1,1,1]])
Suma_kernel = N_kernel + S_kernel + E_kernel + O_kernel + So_kernel + No_kernel + Se_kernel + Ne_kernel

def prewitt(imagen):
    imagen_prewitt= cv2.filter2D(src=imagen,ddepth=0,kernel=Suma_kernel)
    plt.imshow(imagen_prewitt)
    
```

### Importar imagenes

```python
import cv2
import os

# Ruta de la carpeta que contiene las imágenes
carpeta_imagenes = "./Train1"

# Obtener la lista de archivos en la carpeta de imágenes
archivos = os.listdir(carpeta_imagenes)

# Filtrar solo los archivos de imágenes (puedes ajustar esto según los formatos de imagen que desees)
archivos_imagenes = [archivo for archivo in archivos if archivo.endswith(('.jpg', '.jpeg', '.png', '.gif'))]

# Crear una lista para almacenar las imágenes
imagenes = []

# Iterar sobre los archivos de imágenes y leer cada una usando OpenCV
for nombre_imagen in archivos_imagenes:
    ruta_imagen = os.path.join(carpeta_imagenes, nombre_imagen)
    imagen = cv2.imread(ruta_imagen)
    if imagen is not None:
        imagenes.append(imagen)
    else:
        print(f"No se pudo leer la imagen: {nombre_imagen}")

# Ahora la lista "imagenes" contiene todas las imágenes de la carpeta especificada, 
# y puedes procesarlas según tus necesidades.
```

```python
nombre_imagen_seleccionada = random.choice(archivos_imagenes)
ruta_imagen_seleccionada = os.path.join(carpeta_imagenes, nombre_imagen_seleccionada)

# Leer la imagen seleccionada
imagen = cv2.imread(ruta_imagen_seleccionada)

# Aplicar el filtro Canny a la imagen
imagen_canny = cv2.Canny(imagen, 100, 200)  # Ajusta los umbrales según sea necesario

# Configurar la visualización de la imagen original y la imagen con el filtro Canny aplicado
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
fig.subplots_adjust(wspace=0.3)

# Mostrar la imagen original
axs[0].imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
axs[0].axis('off')
axs[0].set_title('Imagen Original')
# Mostrar la imagen con el filtro Canny aplicado
axs[1].imshow(imagen_canny, cmap='gray')
axs[1].axis('off')
axs[1].set_title('Filtro Canny')

plt.suptitle(f'Aplicación del filtro Canny a')
plt.show()


plt.show()
```
