```python
import cv2
from PIL import Image
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import skimage.morphology as morph
```

```python
conos = cv2.imread('Conos.jpg',0)
bicicleta = cv2.imread('Bicicleta.jpg',0)
crosswalk = cv2.imread('Crosswalk.jpg',0)
hidrantes = cv2.imread('Hidrantes.jpg', 0)
```

```python
plt.figure(1)
plt.subplots_adjust(top=1.75,bottom=0.3,left=0.10,right=2,hspace=0.25,wspace=0.5)
plt.subplot(221),plt.imshow(bicicleta,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(conos,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(crosswalk,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(hidrantes,cmap='gray')
plt.xticks([]), plt.yticks([])
```

```python
def filtros (imagen, iteraciones):
    ret, imagen_th = cv2.threshold(imagen, 50, 255, cv2.THRESH_BINARY_INV)
    se1 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,5))
    imagen_th = cv2.dilate(imagen_th,se1,iterations=2)
    imagen_adel = morph.thin(imagen_th, iteraciones)
    plt.figure()
    plt.subplot(131),plt.imshow(imagen,cmap='gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(imagen_adel,cmap='gray')
    plt.title('Adelgazamiento'),plt.xticks([]),plt.yticks([])
    imagen_sk = morph.skeletonize(imagen_th // 255)
    plt.subplot(133),plt.imshow(imagen_sk,cmap='gray')
    plt.title('Esqueleto'),plt.xticks([]),plt.yticks([])
    
    
```

```python
filtros(hidrantes,20)
```

```python
filtros(bicicleta,50)
```

```python
filtros(crosswalk,50)
```

```python
filtros(conos,30)
```

## Adelgazamiento
Al aplicar la funcin de adelgazamiento a imgenes que contienen bicicletas, hidrantes, conos y cruces peatonales, se destacan los detalles finos y los contornos de los objetos, como los componentes individuales de las bicicletas, las partes estructurales de los hidrantes, los bordes de los conos y las lneas de los cruces peatonales. Esto facilita la identificacin y el seguimiento de estos objetos en aplicaciones de seguridad vial, gestin urbana y deteccin de anomalas.

## Esqueleto
La aplicacin de la funcin de esqueleto a estas imgenes permite representar la estructura principal y la disposicin espacial de los objetos, como la forma general de las bicicletas, la ubicacin central de los hidrantes, la orientacin de los conos y la extensin de los cruces peatonales. Esto simplifica la deteccin, localizacin y anlisis de estos elementos en entornos urbanos, contribuyendo a la seguridad vial y la gestin eficiente de la infraestructura urbana.
