```python
import cv2
from PIL import Image
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import skimage.morphology as morph
```

```python
conos = cv2.imread('Conos.jpg',0)
bicicleta = cv2.imread('Bicicleta.jpg',0)
crosswalk = cv2.imread('Crosswalk.jpg',0)
hidrantes = cv2.imread('Hidrantes.jpg', 0)
```

```python
plt.figure(1)
plt.subplots_adjust(top=1.75,bottom=0.3,left=0.10,right=2,hspace=0.25,wspace=0.5)
plt.subplot(221),plt.imshow(bicicleta,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(conos,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(crosswalk,cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(hidrantes,cmap='gray')
plt.xticks([]), plt.yticks([])
```

```python
def filtros (imagen, iteraciones):
    ret, imagen_th = cv2.threshold(imagen, 50, 255, cv2.THRESH_BINARY_INV)
    se1 = cv2.getStructuringElement(cv2.MORPH_RECT, (3,5))
    imagen_th = cv2.dilate(imagen_th,se1,iterations=2)
    imagen_adel = morph.thin(imagen_th, iteraciones)
    plt.figure()
    plt.subplot(131),plt.imshow(imagen,cmap='gray')
    plt.title('Original'), plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(imagen_adel,cmap='gray')
    plt.title('Adelgazamiento'),plt.xticks([]),plt.yticks([])
    imagen_sk = morph.skeletonize(imagen_th // 255)
    plt.subplot(133),plt.imshow(imagen_sk,cmap='gray')
    plt.title('Esqueleto'),plt.xticks([]),plt.yticks([])
    
    
```

```python
filtros(hidrantes,20)
```

```python
filtros(bicicleta,50)
```

```python
filtros(crosswalk,50)
```

```python
filtros(conos,30)
```

## Adelgazamiento
Al aplicar la función de adelgazamiento a imágenes que contienen bicicletas, hidrantes, conos y cruces peatonales, se destacan los detalles finos y los contornos de los objetos, como los componentes individuales de las bicicletas, las partes estructurales de los hidrantes, los bordes de los conos y las líneas de los cruces peatonales. Esto facilita la identificación y el seguimiento de estos objetos en aplicaciones de seguridad vial, gestión urbana y detección de anomalías.

## Esqueleto
La aplicación de la función de esqueleto a estas imágenes permite representar la estructura principal y la disposición espacial de los objetos, como la forma general de las bicicletas, la ubicación central de los hidrantes, la orientación de los conos y la extensión de los cruces peatonales. Esto simplifica la detección, localización y análisis de estos elementos en entornos urbanos, contribuyendo a la seguridad vial y la gestión eficiente de la infraestructura urbana.
